```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize, special
from numpy import loadtxt
```

## 1. Analytical solution of dispersion relation

- Expression:
$$\hat{\omega} =  (\frac{1}{2} + i\frac{\sqrt{3}}{2})\mid(k_y \lambda_{i})^{1/3} \left(\frac{L_{T}}{\lambda_{i}}\right)^{-1/3} \left(\frac{T_e}{T_i}\right)^{-1/3}(k_z \lambda_{i})^{2/3} \mid$$ 

- Considered parameters:
    - $k_y \lambda_{i}=0.1$
    - $\frac{L_{T}}{\lambda_{i}}=10$
    - $\frac{T_e}{T_i}=10$

```python
# calculate analytical solution

# set parameters
k_y_lambda = 0.1 # k_y * lambda_i
L_T_lambda = 10 # L_T / lambda_i 
k_z_lambda = np.logspace(-5, -2, 1000) # k_z * lambda_i
temp_ratio = 10

# calculate omega analytical
omega_analytical = (1/2 - np.sqrt(3)/2j)*np.absolute(k_y_lambda**(1/3) * 
                                          L_T_lambda**(-1/3) * 
                                          temp_ratio**(-1/3) * 
                                          (k_z_lambda)**(2/3))
```

## 2. Load data for the reference figure

```python
# load data for the reference figure
analytical_data_imaginary = loadtxt('D:/MSc/OneDrive - University of York/MSc/project/Analytical_imaginary.csv', delimiter=',')
numerical_data_imaginary = loadtxt('D:/MSc/OneDrive - University of York/MSc/project/Numerical_imaginary.csv', delimiter=',')

analytical_data_real = loadtxt('D:/MSc/OneDrive - University of York/MSc/project/Analytical_real.csv', delimiter=',')
numerical_data_real = loadtxt('D:/MSc/OneDrive - University of York/MSc/project/Numerical_real.csv', delimiter=',')
```

```python
# set lists for the data of the reference figure

# real part analytical
k_z_lambda_analytical_real = []
frequency_analytical_real = []

# imaginary part analytical
k_z_lambda_analytical_imaginary = []
growth_rate_analytical = []

# store the analytical real part
for i in range(len(analytical_data_real)):
    
    k_z_lambda_analytical_real.append(analytical_data_real[i][0])
    frequency_analytical_real.append(analytical_data_real[i][1])
    
# store the analytical imaginary part 
for i in range(len(analytical_data_imaginary)): 
    
    k_z_lambda_analytical_imaginary.append(analytical_data_imaginary[i][0])
    growth_rate_analytical.append(analytical_data_imaginary[i][1])
```

```python
# compare our solution against the reference figure 

# real part
plt.subplots(figsize=(10,5))

# our analytical solution
plt.plot(k_z_lambda, omega_analytical.real, label='Analytical')
# reference analytical solution
plt.plot(k_z_lambda_analytical_real, frequency_analytical_real, label='Analytical_Ref')

plt.xlabel('$k_z \lambda_i$ [normalised]', fontsize=15.0)
plt.title('Comparison of real part of the normalised frequency')
plt.ylabel('Real($\omega / \Omega_i$) [normalised]', fontsize=15.0)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="-")
plt.legend()
plt.show()

# imaginary part
plt.subplots(figsize=(10,5))

# our analytical solution
plt.plot(k_z_lambda, omega_analytical.imag, label='Analytical')
# reference analytical solution
plt.plot(k_z_lambda_analytical_imaginary, growth_rate_analytical, label='Analytical_Ref')

plt.xlabel('$k_z \lambda_i$ [normalised]', fontsize=15.0)
plt.title('Comparison of imaginary part of the normalised frequency')
plt.ylabel('Imag($\omega / \Omega_i$) [normalised]', fontsize=15.0)
plt.xscale('log')
plt.grid(True, which="both", ls="-")
plt.legend()
plt.show()
```

## 3. Change the expression for the analytical solution

- Change the expression to:
$$\hat{\omega} =  (\frac{1}{2} + i\frac{\sqrt{3}}{2})\mid(k_y \lambda_{i})^{1/3} \left(\frac{L_{T}}{\lambda_{i}}\right)^{1/3} \left(\frac{T_e}{T_i}\right)^{-1/3}(k_z \lambda_{i})^{2/3} \mid$$ 

- or:
$$\hat{\omega} =  (\frac{1}{2} + i\frac{\sqrt{3}}{2})\mid(k_y \lambda_{i})^{1/3} \left(\frac{L_{T}}{\lambda_{i}}\right)^{-1/3} \left(\frac{T_e}{T_i}\right)^{1/3}(k_z \lambda_{i})^{2/3} \mid$$ 


- Either expression will give the same result, since these considered parameters ($L_T/ \lambda_{i}, T_e/T_i$) are equal.

- Considered parameters:
    - $k_y \lambda_{i}=0.1$
    - $\frac{L_{T}}{\lambda_{i}}=10$
    - $\frac{T_e}{T_i}=10$

```python
# calculate the new analytical solution

# set parameters
k_y_lambda = 0.1 # k_y * lambda_i
L_T_lambda = 10 # L_T / lambda_i 
k_z_lambda = np.logspace(-5, -2, 1000) # k_z * lambda_i
temp_ratio = 10

# calculate omega analytical
new_omega_analytical = (1/2 - np.sqrt(3)/2j)*np.absolute(k_y_lambda**(1/3) * 
                                          L_T_lambda**(1/3) * 
                                          temp_ratio**(-1/3) * 
                                          (k_z_lambda)**(2/3))
```

## 4. Make the new comparison

```python
# compare the new solution against the reference figure 

# real part
plt.subplots(figsize=(10,5))

# our analytical solution
plt.plot(k_z_lambda, new_omega_analytical.real, label='Analytical')
# reference analytical solution
plt.plot(k_z_lambda_analytical_real, frequency_analytical_real, label='Analytical_Ref')

plt.xlabel('$k_z \lambda_i$ [normalised]', fontsize=15.0)
plt.title('Comparison of real part of the normalised frequency')
plt.ylabel('Real($\omega / \Omega_i$) [normalised]', fontsize=15.0)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="-")
plt.legend()
plt.show()

# imaginary part
plt.subplots(figsize=(10,5))

# our analytical solution
plt.plot(k_z_lambda, new_omega_analytical.imag, label='Analytical')
# reference analytical solution
plt.plot(k_z_lambda_analytical_imaginary, growth_rate_analytical, label='Analytical_Ref')

plt.xlabel('$k_z \lambda_i$ [normalised]', fontsize=15.0)
plt.title('Comparison of imaginary part of the normalised frequency')
plt.ylabel('Imag($\omega / \Omega_i$) [normalised]', fontsize=15.0)
plt.xscale('log')
plt.grid(True, which="both", ls="-")
plt.legend()
plt.show()
```
