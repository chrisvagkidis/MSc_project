## Analytical solution II 

- Expression:
$$\hat{\omega} =  (\frac{1}{2} + i\frac{\sqrt{3}}{2})\mid(k_y \lambda_{i})^{1/3} (\frac{L_{T}}{\lambda_{i}})^{-1/3} (k_z \lambda_{i})^{2/3} \mid$$ 

- Considered parameters:
    - $k_y \lambda_{i}=0.1$
    - $\frac{L_{T}}{\lambda_{i}}=10$

```python
k_y_lambda = 0.1 # k_y * lambda_i
L_T_lambda = 10 # L_T / lambda_i 
k_z_lambda = np.linspace(0.00001, 0.001, 1000)

omega = (1/2 - np.sqrt(3)/2j)*np.absolute(k_y_lambda**(1/3) * L_T_lambda**(-1/3) * (k_z_lambda)**(2/3))

plt.subplots(figsize=(10,5))
plt.plot(k_z_lambda, omega.real)
plt.xlabel('$k_z \lambda_i$', fontsize=15.0)
plt.ylabel('Real($\omega / \Omega_i$)', fontsize=15.0)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="-")
plt.show()

plt.subplots(figsize=(10,5))
plt.plot(k_z_lambda, omega.imag)
plt.xlabel('$k_z \lambda_i$', fontsize=15.0)
plt.ylabel('Imag($\omega / \Omega_i$)', fontsize=15.0)
plt.xscale('log')
plt.grid(True, which="both", ls="-")
plt.show()
```

## Analytical solution III

- Expression:
$$\hat{\omega} =  (\frac{1}{2} + i\frac{\sqrt{3}}{2})\mid(k_y \lambda_{i})^{1/3} \left(\frac{L_{T}}{\lambda_{i}}\right)^{-1/3} \left(\frac{T_e}{T_i}\right)^{-1/3}(k_z \lambda_{i})^{2/3} \mid$$ 

- Considered parameters:
    - $k_y \lambda_{i}=0.1$
    - $\frac{L_{T}}{\lambda_{i}}=10$
    - $\frac{T_e}{T_i}=10$

```python
# normalised solution for the ITG
# plots real part and imaginary part of frequency

import numpy as np
import matplotlib.pyplot as plt

# set parameters
k_y_lambda = 0.1 # k_y * lambda_i
L_T_lambda = 10 # L_T / lambda_i 
k_z_lambda = np.logspace(-5, -2, 1000) # k_z * lambda_i
temp_ratio = 10

# calculate omega
omega = (1/2 - np.sqrt(3)/2j)*np.absolute(k_y_lambda**(1/3) * 
                                          L_T_lambda**(-1/3) * 
                                          temp_ratio**(-1/3) * 
                                          (k_z_lambda)**(2/3))

# plot frequency
plt.subplots(figsize=(10,5))

# plot real part
plt.plot(k_z_lambda, omega.real)
plt.xlabel('$k_z \lambda_i$ [normalised]', fontsize=15.0)
plt.title('Real part of the normalised frequency')
plt.ylabel('Real($\omega / \Omega_i$) [normalised]', fontsize=15.0)
plt.xscale('log')
plt.yscale('log')
plt.grid(True, which="both", ls="-")
plt.show()

# plot imaginary part
plt.subplots(figsize=(10,5))
plt.plot(k_z_lambda, omega.imag)
plt.xlabel('$k_z \lambda_i$ [normalised]', fontsize=15.0)
plt.title('Imaginary part of the normalised frequency')
plt.ylabel('Imag($\omega / \Omega_i$) [normalised]', fontsize=15.0)
plt.xscale('log')
plt.grid(True, which="both", ls="-")
plt.show()
```
